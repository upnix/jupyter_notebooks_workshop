*Solving a system of equations (more than one polynomial at a time)*

Now, ChatGPT tells me:  
> To "solve a system of equations" means to find the values of the variables that satisfy all the equations in the system simultaneously.  

So, what values for $x$ satisfy the polynomials $p_1$ and $p_2$?

````python
# (Cell 3306)
# This ensures we're not rendering math output in LaTeX. We want this
# so we can copy and paste the output we're about to see...
%display plain

# This tells SageMath that 'x' and 'y' aren't Python variables that
# might have values, but instead "math" variables.
var('x y')

# We're asking SageMath to find values for 'x' and 'y' for which both
# polynomials/equations/functions will be satisfied.
solve([p_1(x)==y, p_2(x)==y], x, y)
````

Ohmygoodness. We can interpret this as:
$p_1$ and $p_2$ having two solutions:
1. $x = -\frac{1}{10}\cdot \sqrt{249} - \frac{13}{10}$, $y = \frac{33}{50} \cdot \sqrt{249} + \frac{669}{50}$  

*AND...*  

2. $x = \frac{1}{10} \cdot \sqrt{249} - \frac{13}{10}$, $y = -\frac{33}{50} \cdot \sqrt{249} + \frac{669}{50}$

So, "For both $p_1$ and $p_2$, when $x$ equals *this*, $y$ will equal *that*."

````python
# (Cell 3307)
# Lets grab these solutions and assign them to their own variables so
# it's easier to work with them going forward.

# The whole list of solutions found:
solved_p1p2 = solve([p_1(x)==y, p_2(x)==y], x, y)

# Pulling out just the first solution for 'x'
x_1 = solved_p1p2[0][0].rhs()
# Pulling out just the first solution for 'y'
y_1 = solved_p1p2[0][1].rhs()

# Same deal
x_2 = solved_p1p2[1][0].rhs()
y_2 = solved_p1p2[1][1].rhs()

# What about all that other stuff ("[1][0].rhs()," etc.)? That's a
# little deeper into the Python weeds. `solved_p1p2` is a Python `list()`
# data structure, where, in this case, each item on the list is itself a
# list.
# -> The square brackets: We're "subsetting" the list data. '[0]' means
#    "Give me the very first item from the list."
#    Reading '[0][0]' right-to-left, we're saying: "Give me the first
#      element from the list that is itself the first element from the
#      first list." Yikes.
#
# -> `.rhs()` means "Right hand side." The Python object held in the
#    second list (accessed via `solved_p1p2[a][b]`) is a SageMath
#    "expression," annnd... I guess `.rhs()` is the way you say "no, I
#    just want the value represented on the right-hand side."

print("x_1 =", x_1)
print("y_1 =", y_1)
print('')
print("x_2 =", x_2)
print("y_2 =", y_2)
````

````python
# (Cell 3308)
# What does a graph of `p_1` and `p_2` have to say about this?
# We're providing a list of functions (`[p_1, p_2]`) and specifying that
# we want the x-axis kept between -5 and 5
plot([p_1, p_2], xmin=-5, xmax=5)
````

Ok, we see two points of intersection, and previously we saw two solutions when we used the `solve()` function. But, I don't even have the foggiest idea what decimal numbers $x = -\frac{1}{10}\cdot \sqrt{249} - \frac{13}{10}$, $y = \frac{33}{50} \cdot \sqrt{249} + \frac{669}{50}$ , for example, are. Could those $x$ and $y$ values be even remotely near the intersection points on the graph?  

Lets use the `N()` function. We speed things up by copying the same four `print()` from the code cell 3306, then add in `N()`:

````python
# (Cell 3309)
print("x_1 =", N(x_1))
print("y_1 =", N(y_1))
print('')
print("x_2 =", N(x_2))
print("y_2 =", N(y_2))
````

````python
# (Cell 3310)
# I'm still on the fence. Lets plot these points on the same graph:

# This is the same `plot()` function from before, but now we're using
# SageMath's `point()` function to add points. To put this all on the
# same plot we uses `+`!
plot([p_1, p_2], xmin=-5, xmax=5)+point((x_1,y_1), size=30)+point((x_2,y_2), size=30)
````

````python
# (Cell 342)

# !! This is a cell you can just execute and move on from !! #

# The following code is 95% from here:
# https://medium.com/@mathcube7/how-to-make-interactive-plots-in-jupyter-notebooks-211c4e18321a

# Don't worry too much about what's going on here. If you're worried about
# how complicated this looks I should point out that this is not a SageMath
# method for making interactive charts (See the appendix section
# "Interactive Graphing with SageMath"). SageMath offers a much simpler
# way for turning functions and equations into interactive objects.
#
# The remaining comments will be insufficient for understanding what's
# happening unless you already know a bit about Python.

# Import the Python module 'Numpy'
import numpy as np
# Import the Python module 'IPywidgets'. This is what makes the
# Matplotlib chart interactive.
import ipywidgets as w
# Import the Python module 'matplotlib', this is what makes the chart.
import matplotlib.pyplot as pp

# This Python function is run every time the slider widgets are changed
def update_plot(a, b, c, d):
    # This generates a lot of numbers between -5, 5. They're fed into
    # the polynomial we're plotting in order to get a smooth curve.
    x = np.linspace(-5, 5, 200)
    
    # This is the function/polynomial of interest. Note that instead of
    # using '^' power exponents it's using the Python-standard '**'
    y = a*x**3 + b*x**2 + c*x + d
    
    # Make the plot using Matlibplot
    pp.plot(x, y, linewidth=4)
    
    # Label the plot
    pp.xlabel('x')
    pp.ylabel('f(x)')
    
    # Add a grid
    pp.grid()
    
    # Keep the y-axis fixed between -5 and 5
    pp.ylim(-5, 5)

    
# This allows us to print the formula being graphed, right next to the graph
def update_formula(a, b, c, d):
    # The IPywidget that will display the LaTeX string
    label = w.Label()
    
    # This holds the LaTeX formatted formula, with spots for the variables
    # 'a', 'b', 'c', and 'd'.
    formula = r"$\large{{y = {}x^3 + ({})x^2 + ({})x + {}}}$".format(round(a, 2), round(b, 2), round(c, 2), round(d, 2))
    
    # Give the widget the string it'll show
    label.value = formula
    
    # Displays the label where ever the calling method,
    # `interactive_output()`, decides.
    display(label)

    
# This Python function constructs the interactive UI. It'll only be run
# once, whereas `update_plot()` is run every time you move a slider.
def build_ui():
    # Make 4 sliders, labeled a through d. Only allow them to move between
    # values -3 and 3.
    a, b, c, d = sliders = [
        w.FloatSlider(min=-3, max=3, description=desc)
        for desc in 'abcd'
    ]
    
    # Indicate that the function `update_plot()` is the interactive output
    plot = w.interactive_output(update_plot, {'a':a, 'b':b, 'c':c, 'd':d})
    
    formula = w.interactive_output(update_formula, {'a':a, 'b':b, 'c':c, 'd':d})
    
    # Stack formula and sliders vertically
    controls = w.VBox([formula] + sliders)
    
    # Put the vertically-stacked controls horizontal with the plot
    ui = w.HBox([controls, plot])
    
    # Send the constructed UI to whoever called this function. In our
    # case, that's Jupyter.
    return ui


# Call the function that was defined above to kick things off!
build_ui()
````

### Plotting mutiple countries at once

We'll start by making a simple Python function for creating plots:

````python
# (Cell 465)
# This function takes one country, and one cause of death
def ez_plot(country, cause):
    # Create a DataFrame by pulling out the requested country and cause
    # data
    data = deaths.loc[country][cause]
    
    # Make a plot. Again, the `.index` happens to be the year and
    # `.values` is the corresponding value for that index (year) value.
    pp.plot(data.index, data.values)

print('Just defined ez_plot()!')
````

````python
# (Cell 466)
# Test out our new plotting function
ez_plot('Mexico', 'parkinsons')
````

## 7.1. Is your kernel busy, lazy, or dead?

In the background, executing all your notebook code, is a Python "kernel." We're going to be running some code that can take a very long time to finish - or that may never finish! When your kernel is busy it's unlikely to be creating any output, which can make it hard to tell what's happening. However, you can see if the Python kernel is busy by looking at the kernel status on the lower JupyterLab status bar. It'll look like this: 

If it says "Idle," then there's some other reason you're not seeing any output. If it says "Busy" you can wait some more, or kill the kernel by pressing the "Stop" button on the toolbar, above.

### 7.2. Restart your kernel

Jupyter has what is called a "kernel," which is the background code interpreter that executes code from your notebook cells - we're using "Python 3" as a kernel. It's often the case that we execute some code that made our Python kernel unhappy, or, that put our environment into a state that's interfering with new code we're trying to execute. To start from a clean slate, restart the Python kernel by clicking the little "reload" round-arrow-thing in the toolbar above.

## A.5. Shell commands via Jupyter

***What's up with the `!` commands?***  
This is a Jupyter-specific convention (so, not Python or any other programming language) that causes the command following the `!` to be run on the *system shell*. Think of the shell as a portal to the "behind the scenes" machinery that's allowing Jupyter to run.

We don't need to understand this to work with Jupyter, so don't worry too much. See (Appendix) if you do want to learn more. As an overview, the various systems involved in allowing JupyterLab to run for you are:

The shell you're accessing with the `!` convention is pretty much right at (2.5).
